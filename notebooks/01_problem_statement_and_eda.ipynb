````python
# Check for outliers using IQR method
def find_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    n_outliers = len(data[(data[column] < lower_bound) | (data[column] > upper_bound)])
    return n_outliers, lower_bound, upper_bound

# Get current numerical columns
current_numerical = df_clean.select_dtypes(include=[np.number]).columns.tolist()

print("Outlier analysis:")
print("="*60)
for col in current_numerical:
    n_outliers, lower, upper = find_outliers_iqr(df_clean, col)
    if n_outliers > 0:
        pct = n_outliers / len(df_clean) * 100
        print(f"{col}: {n_outliers:,} outliers ({pct:.1f}%) | bounds: [{lower:.2f}, {upper:.2f}]")
````

````python
# YOUR ADDITIONAL FEATURE ENGINEERING CODE HERE
# Add as many cells as you need!

# Feature 4: Peak Season Flag (Domain Knowledge)
# I extract the month number from the date to identify high-demand periods.
# Strategy: Dec(12), Jan(1) are holidays. May(5)-July(7) are summer vacation.

temp_month = df_clean['Travel Date'].dt.month
peak_months = [12, 1, 5, 6, 7] 

df_clean['Is_Peak_Season'] = temp_month.apply(lambda x: 1 if x in peak_months else 0)

# Feature 5: Interaction Feature (Trip Complexity)
# I combine 'Num_Places_Visited' and 'Flight Stops' to create a "Logistical Difficulty" score.
# Logic: More stops AND more cities = Higher Price.
# We add +1 to Flight Stops to avoid multiplying by zero.
df_clean['Trip_Complexity'] = df_clean['Num_Places_Visited'] * (df_clean['Flight Stops'] + 1)

print("Additional Engineering Complete!")
print("-" * 30)
print(df_clean[['Travel Date', 'Is_Peak_Season', 'Flight Stops', 'Num_Places_Visited', 'Trip_Complexity']].head())
print(f"\nFinal Shape: {df_clean.shape}")
````

**Explain your additional feature engineering:**

I created two domain-specific features to capture market behavior and logistical costs:

Is_Peak_Season (Binary):

Logic: Travel prices are highly seasonal and surge during holidays and summer breaks due to high demand.

Implementation: I created a binary flag (1/0) derived from the Travel Date. Months 12, 1 (Winter Holidays) and 5, 6, 7 (Summer Vacation) are marked as 1 (Peak), while others are 0. This helps the model differentiate between "expensive by nature" periods and regular dates.

Trip_Complexity (Interaction Feature):

Logic: A simple count of cities or flights doesn't tell the whole story. A trip covering 5 cities and requiring 2 flight stops is exponentially more complex logistically than a direct flight to a single city.

Implementation: I calculated Num_Places_Visited * (Flight Stops + 1). This interaction term highlights complex, multi-leg journeys which typically command a higher premium price.
